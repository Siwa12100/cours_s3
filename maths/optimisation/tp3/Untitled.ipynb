```python
# ============ 
# Exercice 1 :
# ============

def tournee_gloutonne(M):
    n = len(M)
    T = [0]
    non_visitees = set(range(1, n))  

    while non_visitees:
        i = T[-1]
        # Distance min... : 
        j = min(non_visitees, key=lambda x: M[i][x])  
        # On l'ajoute... : 
        T.append(j)
        # On l'enlve des non visits 
        non_visitees.remove(j)  

    T.append(T[0]) 
    return T


T = tournee_gloutonne(distances_5villes)
afficher_tournee(T, coordonnees_5villes, noms_5villes, 10)

    
```

```python
# ============
# Exercice 2 : 
# ============

from random import shuffle

def eval(T, M):
    t = 0
    for i in range(len(T) - 1):
        t += M[T[i]][T[i + 1]]
    return t

def echange(T, i, j):
    T_copy = T.copy()
    T_copy[i], T_copy[j] = T_copy[j], T_copy[i]
    return T_copy

def voisinage(T):
    n = len(T) - 1  # Ignorer la ville de dpart  la fin...
    voisins = []

    for i in range(n):
        for j in range(i + 1, n):
            voisin = echange(T, i, j)
            voisins.append(voisin)

    return voisins

def tournee_aleatoire(n):
    tournee = list(range(n))
    shuffle(tournee)
    return tournee

def DescenteGradient(M):
    n = len(M)

    # Solution initiale
    T0 = tournee_aleatoire(n)

    T = T0
    meilleurTaille = eval(T, M)

    # Boucle de descente de gradient
    while True:
        voisins = voisinage(T)

        # Chercher le meilleur voisin
        meilleurTrouve = False
        
        for voisin in voisins:
            taille = eval(voisin, M)
            if taille < meilleurTaille :
                T = voisin
                meilleurTaille = taille
                meilleurTrouve = True

        # Sortir de la boucle si aucun voisin amliorant n'est trouv
        if not meilleurTrouve:
            break

    return T

# Test de la fonction avec distances_5villes
T_optimal = DescenteGradient(distances_48villes)
afficher_tournee(T_optimal, coordonnees_48villes, noms_48villes, 5)
```
